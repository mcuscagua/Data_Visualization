```python
import pandas as pd
import numpy as np
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go
from plotly.subplots import make_subplots
init_notebook_mode()
```

```python
precios = pd.read_excel('PX.xlsx')
precios = precios[precios['Dates'] >= pd.datetime(2015,1,1)]
precios.columns = ['Dates', 'SPY', 'IWM', 'EZU']
precios['SPY'] = np.cumsum(precios['SPY']/precios['SPY'].shift(1) - 1)
precios['IWM'] = np.cumsum(precios['IWM']/precios['IWM'].shift(1) - 1)
precios['EZU'] = np.cumsum(precios['EZU']/precios['EZU'].shift(1) - 1)

```

```python
precios
```

```python
precios100 = precios.copy()
precios100['SPY'] = precios100['SPY']*100
precios100['IWM'] = precios100['IWM']*100
precios100['Spread'] = precios100['SPY'] - precios100['IWM']
```

```python
fig = make_subplots(rows=2, cols=1, start_cell="bottom-left", shared_xaxes=True)

fig.add_trace(go.Scatter(y=precios100['SPY'],x=precios100['Dates'],marker=dict(color='blue'),mode='lines', name='SPY'),
              row=2, col=1)
fig.add_trace(go.Scatter(y=precios100['IWM'],x=precios100['Dates'],marker=dict(color='red'),mode='lines', name='IWM'),
              row=2, col=1)
fig.add_trace(go.Scatter(y=precios100['Spread'],x=precios100['Dates'],marker=dict(color='green'),mode='lines', name='Diferencia de las series'),
              row=1, col=1)

fig.add_trace(go.Scatter(y=[0 for i in range(precios100.shape[0])],x=precios100['Dates'],marker=dict(color='black'),mode='lines', name='Referente de reversion a la media'),
              row=1, col=1)

fig.update_layout(title='Oportunidad de Arbitraje de Indices con referente base 100 desde Enero 2015',
#                   xaxis=dict(title="Tiempo"),
                  yaxis=dict(title="Diferencia SPY - IWM"),
                  yaxis2=dict(title="Retorno Acumulado Base 100"))

fig.show()
```

```python
fig.write_html('Grafica.html')
```
